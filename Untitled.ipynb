```python
# Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data from URL using pandas read_csv method
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/00616/Tetuan%20City%20power%20consumption.csv')
```

```python
df.head()
```

```python
print(df.columns)
```

```python
# Convert DateTime column to datetime format
df['DateTime'] = pd.to_datetime(df['DateTime'])
```

```python
# Set DateTime as index
df.set_index('DateTime', inplace=True)
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Check for duplicate values
print(df.duplicated().sum())
```

```python
# Summary statistics
print(df.describe())
```

```python
# EDA - Line plot of power consumption over time
plt.figure(figsize=(15,5))
plt.plot(df.index, df['Zone 1 Power Consumption'], label='Zone 1 Power Consumption')
plt.plot(df.index, df['Zone 2  Power Consumption'], label='Zone 2 Power Consumption')
plt.plot(df.index, df['Zone 3  Power Consumption'], label='Zone 3 Power Consumption')
plt.xlabel('Time')
plt.ylabel('Power Consumption (KW)')
plt.title('Power Consumption over Time')
plt.legend()
plt.show()
```

```python
# EDA - Heatmap of correlation matrix
corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
```

```python
# Split data into training and testing sets
X = df[['Temperature','Humidity','Wind Speed','general diffuse flows','diffuse flows']]
y1 = df['Zone 1 Power Consumption']
y2 = df['Zone 2  Power Consumption']
y3 = df['Zone 3  Power Consumption']
from sklearn.model_selection import train_test_split
X_train, X_test, y1_train, y1_test, y2_train, y2_test, y3_train, y3_test = train_test_split(X, y1, y2, y3, test_size=0.2, random_state=42)
```

```python
# Train and evaluate regression models
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Linear Regression
linreg1 = LinearRegression()
linreg1.fit(X_train, y1_train)
linreg1_y_pred = linreg1.predict(X_test)
print('Zone 1 Linear Regression')
print('Mean Squared Error: ', mean_squared_error(y1_test, linreg1_y_pred))
print('R2 Score: ', r2_score(y1_test, linreg1_y_pred))
```

```python
linreg2 = LinearRegression()
linreg2.fit(X_train, y2_train)
linreg2_y_pred = linreg2.predict(X_test)
print('Zone 2 Linear Regression')
print('Mean Squared Error: ', mean_squared_error(y2_test, linreg2_y_pred))
print('R2 Score: ', r2_score(y2_test, linreg2_y_pred))
```

```python
linreg3 = LinearRegression()
linreg3.fit(X_train, y3_train)
linreg3_y_pred = linreg3.predict(X_test)
print('Zone 3 Linear Regression')
print('Mean Squared Error: ', mean_squared_error(y3_test, linreg3_y_pred))
print('R2 Score: ', r2_score(y3_test, linreg3_y_pred))
```

```python
# Train XGBoost model
import xgboost as xgb
xgb_model = xgb.XGBRegressor(objective='reg:squarederror', n_estimators=100, max_depth=3, learning_rate=0.1)
xgb_model.fit(X_train, y1_train)
```

```python
# Make predictions on test set
y_pred = xgb_model.predict(X_test)
```

```python
# Create a dataframe to store actual and predicted values
results_df = pd.DataFrame({'Actual': y1_test, 'Predicted': y_pred})
```

```python
# Plot actual vs predicted values
plt.figure(figsize=(10, 6))
plt.plot(results_df.index, results_df['Actual'], label='Actual')
plt.plot(results_df.index, results_df['Predicted'], label='Predicted')
plt.legend()
plt.title('Actual vs Predicted Zone 1 Power Consumption')
plt.xlabel('Observation')
plt.ylabel('Power Consumption (KW)')
plt.show()
```
